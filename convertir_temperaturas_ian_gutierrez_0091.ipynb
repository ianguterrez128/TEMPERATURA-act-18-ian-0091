## 1. Fórmulas de Conversión

Para convertir entre escalas térmicas, utilizamos las siguientes relaciones matemáticas:

* **De Celsius a Fahrenheit:** Multiplicamos los grados Celsius por 1.8 (o $9/5$) y sumamos 32.
    $$°F = (°C \times 1.8) + 32$$

* **De Fahrenheit a Celsius:** Restamos 32 a los grados Fahrenheit y luego dividimos el resultado entre 1.8.
    $$°C = \frac{°F - 32}{1.8}$$



---

## 2. Ejemplos Resueltos Paso a Paso

### Ejemplo A: Convertir 25°C a Fahrenheit (Temperatura ambiente cómoda)
1.  **Identificar la fórmula:** $°F = (°C \times 1.8) + 32$
2.  **Sustituir el valor:** $°F = (25 \times 1.8) + 32$
3.  **Multiplicar:** $25 \times 1.8 = 45$
4.  **Sumar 32:** $45 + 32 = 77$
* **Resultado:** 25°C equivalen a **77°F**.

### Ejemplo B: Convertir 98.6°F a Celsius (Temperatura corporal promedio)
1.  **Identificar la fórmula:** $°C = (°F - 32) / 1.8$
2.  **Sustituir el valor:** $°C = (98.6 - 32) / 1.8$
3.  **Restar 32:** $98.6 - 32 = 66.6$
4.  **Dividir entre 1.8:** $66.6 / 1.8 = 37$
* **Resultado:** 98.6°F equivalen a **37°C**.

## 3. Código en Python

Este programa utiliza funciones para separar la lógica del cálculo de la interacción con el usuario, tal como lo solicitaste.


### Explicación del Código:
* **`float()`**: Se usa para permitir números con decimales (ej. 36.5).
* **`while True`**: Crea un bucle infinito que solo se detiene cuando el usuario elige la opción 3.
* **`f"{valor:.2f}"`**: Es un formato para que el resultado solo muestre 2 decimales y sea más fácil de leer.
* **`break`**: Rompe el ciclo `while` inmediatamente, cerrando el programa.


```python
# funcion para convertir de farenheit en celcius
def convertirCF():
    celsius = float(input("\nIngresa los grados Celsius: "))
    fahrenheit = (celsius * 1.8) + 32
    print(f"--- {celsius}°C es igual a {fahrenheit:.2f}°F ---")

# Función para convertir de Fahrenheit a Celsius
def convertirFC():
    fahrenheit = float(input("\nIngresa los grados Fahrenheit: "))
    celsius = (fahrenheit - 32) / 1.8
    print(f"--- {fahrenheit}°F es igual a {celsius:.2f}°C ---")

# Función principal del menú
def menu():
    while True:
        print("\n--- MENÚ DE CONVERSIONES ---")
        print("1. Convertir grados Centígrados a Fahrenheit")
        print("2. Convertir grados Fahrenheit a Centígrados")
        print("3. Salir")

        opcion = input("\nSelecciona una opción (1-3): ")

        if opcion == '1':
            convertirCF()
        elif opcion == '2':
            convertirFC()
        elif opcion == '3':
            print("Saliendo del programa... ¡Hasta luego!")
            print("¡ian gutierrez NL = 22")
            break
        else:
            print("Opción no válida, por favor intenta de nuevo.")

# Llamada al programa
menu()
```
